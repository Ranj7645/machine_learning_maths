A simple line 

y = 2X+1

```python
def line_func(X):
    return 2 *X + 1
```

```python
x = np.linspace(-3,3,100)
y = line_func(x)
```

```python
print(x[:5])  # Print the first 5 values of x
print(y[:5])  # Print the first 5 values of y
```

```python
# Given x value and y valuen find the slo[pe of the line

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")    
plt.ylabel("Y-axis")
plt.grid(alpha=0.3)
plt.show()

x1, x2 =-2,1
y1, y2 = line_func(x1), line_func(x2)
slope = (y2 - y1) / (x2 - x1)
print(f"Slope of the line between points ({x1}, {y1}) and ({x2}, {y2}) is: {slope}")

```

----------------- Curve x^2 ------------------------------

```python
def curve_func(X):
    return X**2

x = np.linspace(-3,3,100)
y =curve_func(x)

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title("Curve y = x^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(alpha=0.3)
plt.show()
```

```python
def slope_estimate(func,x1,h=0.0001):
    x2 = x1 + h
    y1,y2 = func(x1), func(x2)
    # slope dy/dx = (y2 - y1) / (x2 - x1)
    dy = y2 - y1
    dx = h
    slope = dy / dx
    return slope
```

```python
points = [-2, 0, 2]

for p in points:
    s = slope_estimate(curve_func, p)
    print(f"slope of curve for x :{p} is {s:.2f}")

```

```python
x = np.array([1, 2, 3])
y_true = np.array([2, 4, 6])

#  starting guess of w
w = 1.0

# learning rate
lr = 0.01
ws = list()
ls = list()
def loss(w):
    y_pred = w*x
    return np.mean((y_pred - y_true)**2)

# dloss -> dL/dw
def dloss(w):
    y_pred = w*x
    return np.mean(2 * x * (y_pred - y_true))


print(f"start weight {w} and loss is {loss(w):.2f}")

for step in range(30):
    slope = dloss(w)
    w = w - lr * slope
    ws.append(w)
    ls.append(loss(w))

    print(f"step {step} w : {round(w)} slope: {slope} loss : {round(loss(w),3)}")

ws_plot = np.linspace(0,4,200)
ls_plot = [loss(wp) for wp in ws_plot]

plt.figure(figsize=(6,4))
plt.plot(ws_plot,ls_plot,label="loss curve")
plt.scatter(ws,ls,color="red", label = "steps of gradient decent")
plt.plot(ws,ls,color="red")
plt.xlabel("Weight")
plt.ylabel("loss value")
plt.title("Gradient descent loss curve ")
plt.legend()
plt.show()
```
