3. measure of skewness

```python
col = df.total_bill
mean_val = np.mean(col)
meadia_val = np.median(col)
mode_val = col.mode()[0]

plt.figure(figsize=(5,3))
sns.histplot(col,kde=True,color="skyblue")
plt.axvline(mean_val, color="red", linestyle="--", label=f"Mean: {mean_val:.2f}")
plt.axvline(meadia_val, color ="green",linestyle= "-",label=f"Median : {meadia_val:.2f}")
plt.legend()
plt.title("distribution of total_bill column")
plt.show()
```

4. measure of spread

spread tells us how scattered data is around the center

```python
# Range
data_min = np.min(col)
print(f"Minimum value: {data_min}")
data_max = np.max(col)
print(f"Maximum value: {data_max}")

data_range = np.ptp(col)
data_range
```

```python
# variance
variance = np.var(col)
print(f"Variance of total_bill column: {variance:.2f}")
```

```python
# standard deviation
std_dev = np.std(col)
print(f"Standard Deviation of total_bill column: {std_dev:.2f}")
```

```python
# IQR (Interquartile Range)
Q1 = np.percentile(col, 25)
Q3 = np.percentile(col, 75)
median_val = np.percentile(col, 50)
print(f"Q2 50th percentile (Median): {median_val:.2f}")
print(f"Q2 50 percentile: {np.percentile(col, 50):.2f}")
IQR = Q3 - Q1
print(f"Interquartile Range (IQR) of total_bill column: {IQR:.2f}")
```

```python
print(df["tip"].skew())
print(df["total_bill"].skew())
```

```python
plt.figure(figsize=(6,3))

plt.subplot(1,2,1)
sns.histplot(df["tip"],kde = True, color = "orange")
plt.title(f"Distribution of tip column {df['tip'].skew():.2f}")

# plot 2
plt.subplot(1,2,2)
sns.histplot(df["total_bill"],kde = True, color = "purple")
plt.title(f"Distribution of total_bill column {df['total_bill'].skew():.2f}")
plt.tight_layout()
plt.show()


print("Interpretation:")
print("Skewness = 0: Perfectly Normal")
print("Skewness > 0: Right Skewed (Tail is on the right)")
print("Skewness < 0: Left Skewed (Tail is on the left)")
```

```python
plt.figure(figsize = (6,3))
sns.boxplot(x = df["total_bill"], color = "lightgreen")
plt.title("Boxplot of total_bill column")
plt.show()


Q1 = np.percentile(df["total_bill"], 25)
Q3 = np.percentile(df["total_bill"], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["total_bill"] < lower_bound) | (df["total_bill"] > upper_bound)]
print(outliers)
print(f"length of outliers: {len(outliers)}")

```

```python
df["total_bill"] > upper_bound
```

```python
# calculate the correlation matrix
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
print(correlation_matrix)

plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()

print(f"\n Correlation between total_bill and tip:  {   correlation_matrix.loc['total_bill', 'tip']:.2f}")
```
