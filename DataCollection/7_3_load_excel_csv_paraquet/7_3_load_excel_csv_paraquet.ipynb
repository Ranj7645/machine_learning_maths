Compare file size

```python
def file_size(path):
    size_bytes = os.path.getsize(path)
    size_kb = round(size_bytes/1024,2)
    size_mb = round(size_bytes/(1024 * 1024),4)
    return size_kb,size_mb

for f in ["titanic.csv","titanic.parquet","titanic.xlsx","titanic_tab.csv"]:
    kb,mb = file_size(f)
    print(f"{f} : {kb}kb and {mb}mb")
```
