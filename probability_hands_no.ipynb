```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# coint toss Simulation

n = 500
outcomes = np.random.choice(["H","T"],size=n)
# print(len(outcomes))
outcomes_head =  np.sum(outcomes == "H")
outcomes_tail = np.sum(outcomes == "T")

print(f" number of heads: {outcomes_head}")
print(f"number of tails is : {outcomes_tail}")

print(f"estimated head is P(H)/P(n) {(outcomes_head/n):.2f}")
print(f"estimated tails is P(T)/P(n) {(outcomes_tail/n):.2f}")
```

```python
# Dice roll simulation

n=1000
dice = np.random.randint(1,7,size=n)
p_four = np.mean(dice == 4)
print(f"Estimated probability of p(4) : {p_four}")
vals, counts = np.unique(dice ,return_counts=True )
print(vals)
print(counts)
plt.figure(figsize=(6,4))
plt.title("Dice roll outcomes")
plt.xlabel("Number")
plt.ylabel("Count")
plt.bar(vals,counts)
plt.show()
```

```python
# Conditional probability
'''
Question :

We create a tiny dataset of 100 students.
Each student may like ML or not.
If they like ML, they are more likely to like DL as well.
Then we compute:
P(ML), P(DL), P(DL | ML)
'''

data = pd.DataFrame({
    "likes_ml" : np.random.choice( [ 1 , 0 ], size = 100, p = [0.6 , 0.4])
})

# DL depends on whether students like ML
data["likes_DL"] = [
    np.random.choice([1,0],p=[0.7,0.3]) if ml else
    np.random.choice([1,0],p=[0.2,0.8])
    for ml in data["likes_ml"]
]
p_ml = data["likes_ml"].mean()
p_dl = data["likes_DL"].mean()
p_dl_given_ml = data[data["likes_ml"] == 1]["likes_DL"].mean()
print(f"probability of liking ML {p_ml}")
print(f"probability of liking DL {p_dl}")
print(f"P(DL|ML):",round(p_dl_given_ml,2))
print(data.head())
```
